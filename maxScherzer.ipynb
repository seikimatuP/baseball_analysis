```python
######
import pandas as pd #  pandas 
import matplotlib.pyplot as plt
from pandas import plotting
import numpy as np

df = pd.read_csv("maxScherzer.csv") 

#df.head()
#df.head(50).style.bar(subset=[])

#df.loc[:,["pitch_name","pfx_x","pfx_z"]]
#df.query("pitch_type == 'FF'").loc[:,["pitch_type","pfx_x","pfx_z"]]

#4
FF_pfx_x = df.query("pitch_name == '4-Seam Fastball'").loc[:,["pfx_x"]] * 30.48
FF_pfx_z = df.query("pitch_name == '4-Seam Fastball'").loc[:,["pfx_z"]] * 30.48
#
CH_pfx_x = df.query("pitch_name == 'Changeup'").loc[:,["pfx_x"]] * 30.48
CH_pfx_z = df.query("pitch_name == 'Changeup'").loc[:,["pfx_z"]] * 30.48
#
CU_pfx_x = df.query("pitch_name == 'Curveball'").loc[:,["pfx_x"]] * 30.48
CU_pfx_z = df.query("pitch_name == 'Curveball'").loc[:,["pfx_z"]] * 30.48
#
SL_pfx_x = df.query("pitch_name == 'Slider'").loc[:,["pfx_x"]] * 30.48
SL_pfx_z = df.query("pitch_name == 'Slider'").loc[:,["pfx_z"]] * 30.48
#
FC_pfx_x = df.query("pitch_name == 'Cutter'").loc[:,["pfx_x"]] * 30.48
FC_pfx_z = df.query("pitch_name == 'Cutter'").loc[:,["pfx_z"]] * 30.48

plt.title("Max Scherxer Pitch Movement")
plt.xlabel("pfx_x")
plt.ylabel("pfx_z")
plt.grid(True)

plt.scatter(FF_pfx_x, FF_pfx_z)
plt.scatter(CH_pfx_x, CH_pfx_z)
plt.scatter(CU_pfx_x, CU_pfx_z)
plt.scatter(SL_pfx_x, SL_pfx_z)
plt.scatter(FC_pfx_x, FC_pfx_z)




```
